```python
#Se analizara los datos de las marcas top en bogota segun su estrato
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
#lectura de los dataframes de marcas reconocidas
path='./LugaresTopBogota-V2.csv'
dfLugaresTop=pd.read_csv(path)
dfLugaresTop=dfLugaresTop.drop(columns='Unnamed: 0')
```

```python
dfLugaresTop
```

```python
dfLugaresTop.describe()
```

```python
#Agrupando por estrato
dfGroupEstrato=dfLugaresTop.groupby(['Estrato']).count()
dfGroupEstrato['categories']=100*dfGroupEstrato['name']/(dfGroupEstrato['name'].sum())
dfGroupEstrato
```

```python
#Agrupando por categoria
dfGroupCategory=dfLugaresTop.groupby(['categories']).count()
dfGroupCategory['categories']=100*dfGroupCategory['name']/(dfGroupCategory['name'].sum())
dfGroupCategory=dfGroupCategory.sort_values(by=['name'],ascending=False)
dfGroupCategory
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#fig, ax=plt.subplots()
labels=['Stratum 2','Stratum 3','Stratum 4','Stratum 5','Stratum 6']
ax.bar(labels,dfGroupEstrato['name'],color='orange')
plt.title('Number of international brand locations in each stratum of Bogota')
plt.ylabel('Number of venues')

for index,value in enumerate(dfGroupEstrato['categories']):
    plt.annotate(str(round(value,1))+'%', xy=(index-0.15,dfGroupEstrato.iloc[index]['name']-7), color='black')
    

plt.savefig('NLugaresTopVSEstrato.pdf', bbox_inches='tight')
plt.savefig('NLugaresTopVSEstrato.png', bbox_inches='tight')
plt.show()
plt.close(fig)

#dfGroupEstrato.plot(kind='bar',y='name')
```

```python
#agrupando por marca y contando
dfGroupMarca=dfLugaresTop.groupby(['MarcaEncontrada']).count()
#dfGroupMarca=dfGroupMarca.sort_values(by=['name'])#,ascending=False)
dfGroupMarca
```

```python
#el 25% de marcas tiene mas de 12.5 lugares en bogota, analizaremos esos marcas que tienen mas lugares 
dfGroupMarcaQ3=dfGroupMarca[dfGroupMarca['name']>=12.5]
dfGroupMarcaQ3=dfGroupMarcaQ3.rename(columns={'name':'NLugares','categories':'Estrato2','lat':'Estrato3','lng':'Estrato4'
                                              ,'id':'Estrato5','NumPunto':'Estrato6'})
dfGroupMarcaQ3=dfGroupMarcaQ3.drop(columns=['Estrato'])
dfGroupMarcaQ3.iloc[:,1:6]=0
dfGroupMarcaQ3
```

```python
pd.set_option('max_rows', None)
```

```python
dfGrpMarcaEstrato=dfLugaresTop.groupby(['MarcaEncontrada','Estrato'], as_index=False).count()
```

```python
#Se llena el df GroupMarcaQ3 con la cantidad de lugares que tiene en cada estrato
for i in dfGroupMarcaQ3.iterrows():
    dfAux=dfGrpMarcaEstrato[dfGrpMarcaEstrato['MarcaEncontrada']==i[0]]
    #print(dfAux)
    #print('_________________________________________')
    for j in range(len(dfAux)):
        dfGroupMarcaQ3.at[i[0], 'Estrato'+str(dfAux.iloc[j][1])]=dfAux.iloc[j][2]
```

```python
'''dfGroupMarcaQ3['Estrato2%']=dfGroupMarcaQ3['Estrato2']*(100)/dfGroupMarcaQ3['NLugares']
dfGroupMarcaQ3['Estrato3%']=dfGroupMarcaQ3['Estrato3']*(100)/dfGroupMarcaQ3['NLugares']
dfGroupMarcaQ3['Estrato4%']=dfGroupMarcaQ3['Estrato4']*(100)/dfGroupMarcaQ3['NLugares']
dfGroupMarcaQ3['Estrato5%']=dfGroupMarcaQ3['Estrato5']*(100)/dfGroupMarcaQ3['NLugares']
dfGroupMarcaQ3['Estrato6%']=dfGroupMarcaQ3['Estrato6']*(100)/dfGroupMarcaQ3['NLugares']'''
dfGroupMarcaQ3
```

```python

labels=['Stratum 2','Stratum 3','Stratum 4','Stratum 5','Stratum 6']

x = 12*np.arange(len(labels))  # the label locations
y = np.arange(0,45,5)
width = 1.5  # the width of the bars

fig1, ax = plt.subplots()

rects1= ax.bar(x - 3*width, dfGroupMarcaQ3.iloc[0][1:6].tolist(), width, label=dfGroupMarcaQ3.index[0])
rects2= ax.bar(x - 2*width, dfGroupMarcaQ3.iloc[1][1:6].tolist(), width, label=dfGroupMarcaQ3.index[1])
rects3= ax.bar(x - 1*width, dfGroupMarcaQ3.iloc[2][1:6].tolist(), width, label=dfGroupMarcaQ3.index[2])
rects4= ax.bar(x          , dfGroupMarcaQ3.iloc[3][1:6].tolist(), width, label=dfGroupMarcaQ3.index[3])
rects5= ax.bar(x + 1*width, dfGroupMarcaQ3.iloc[4][1:6].tolist(), width, label=dfGroupMarcaQ3.index[4])
rects6= ax.bar(x + 2*width, dfGroupMarcaQ3.iloc[5][1:6].tolist(), width, label=dfGroupMarcaQ3.index[5])
rects7= ax.bar(x + 3*width, dfGroupMarcaQ3.iloc[6][1:6].tolist(), width, label=dfGroupMarcaQ3.index[6])


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Venues')
ax.set_title('Number of places in each stratum of the most numerous brands')
ax.set_xticks(x)
ax.set_yticks(y)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects,i):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        j=dfGroupMarcaQ3.iloc[i][0]
        height = rect.get_height()
        ax.annotate('{}%'.format(round(height*100/j,1)),
                    xy=(rect.get_x() + rect.get_width() / 2, height+2),
                    xytext=(0, -15),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        

autolabel(rects1,0)
autolabel(rects2,1)
autolabel(rects3,2)
autolabel(rects4,3)
autolabel(rects5,4)
autolabel(rects6,5)
autolabel(rects7,6)


#fig.tight_layout()
fig1.set_size_inches(30, 5, forward=True)
plt.savefig('NLugaresTopQ3VSEstrato.pdf', bbox_inches='tight')
plt.savefig('NLugaresTopQ3VSEstrato.png', bbox_inches='tight')
plt.show()
plt.close(fig)
```

```python
#dfGroupMarcaQ3Total=dfGroupMarcaQ3.append(dfGroupMarcaQ3.sum(axis=0),ignore_index=False)
#dfGroupMarcaQ3Total

NuevaFila=dfGroupMarcaQ3.sum(axis=0)
NuevaFila.name='Total'
dfGroupMarcaQ3Total=dfGroupMarcaQ3.append(NuevaFila)
dfGroupMarcaQ3Total
```
