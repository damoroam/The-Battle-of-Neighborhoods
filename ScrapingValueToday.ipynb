```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PagIniVT = "https://www.value.today/world/world-top-1000-companies"
PagInitVTReq=requests.get(PagIniVT)
print(PagInitVTReq.status_code)
```

```python
#funcion que extrae la lista de compa√±ias de la pagina ValueToday
def ListValueToday(url):
    #peticion y lectura de la pagina
    print(url)
    PaginaReq=requests.get(url)
    src = PaginaReq.content
    soup = BeautifulSoup(src, 'lxml')
    
    #identificando los segmentos donde estan los nombres de la empresas
    h2=soup.find_all('h2')
    #print(h2)
    
    #extraccion de los nombres de las empresas
    for h2_tag in soup.find_all('h2'):
        a_tag=h2_tag.find('a')
        if a_tag is not None:
            Company.append(a_tag.text)
    #print(Company)        
    
    #extraemos el link de la siguiente pagina
    SigPagina=soup.findAll('li',class_='pager__item')[-2].find('a')
    
    #condicional para la recursividad hasta acabar las paginas
    if SigPagina.find('span').text=='Next page':
        print('pagina')
        linkSigPagparcial=SigPagina.attrs['href']
        Linkcompleto=PagIniVT+linkSigPagparcial
        #print(Linkcompleto)
        ListValueToday(Linkcompleto)
    else:
        print('no hay mas paginas')
    return Company
```

```python
Company=[]
VTList=ListValueToday(PagIniVT)
```
